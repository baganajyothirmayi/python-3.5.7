```python
import matplotlib.pyplot as ps
l=[1,2,3,4,5]
l2=[10,20,30,40,50]
ps.plot(l,l2)
ps.show()
```

```python
import matplotlib.pyplot as ps
l=[1,2,3,4,5]
l2=[10,20,30,40,50]
ps.scatter(l,l2,color="black",marker="^")
ps.plot(l,l2)
ps.xlabel("this is x axis")
ps.ylabel("this is y axis")
ps.title("this is plot diagram")
ps.show()
```

```python
import matplotlib.pyplot as ps
import numpy as np
l=np.array([1,2,3,4,5])
l1=l**2
ps.subplot(2,2,1)
ps.plot(l,l1)
ps.subplot(2,2,2)
ps.plot(l,l1)
ps.subplot(2,2,3)
ps.plot(l,l1)
ps.subplot(2,2,4)
ps.plot(l,l1)
ps.show()
```
