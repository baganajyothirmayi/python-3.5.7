```python
import pandas as pd
def fun_one():
    df=pd.read_csv("emp.csv",index_col="name")
    return df
fun_one()
```

```python
def fun_two():
    df=fun_one()
    df=df.drop('empid',axis=1)
    return df
fun_two()
    
```

```python
def fun_3():
    t=fun_two()
    t1=t.assign(location=["hyd","vizag","mysore","banglore","hyd"])
    t2=t1.assign(role=["frontend","backend",'backend','frontend','frontend'])
    #t3=t2.assign["mouni","salary":50000,"location":"anakapalli","role":"backend"])
    return t2
fun_3()
```

```python
def fun4():
    a=fun_3()
    a=a.sort_values(by='salary',ascending=True)
    return a
fun4()
```

```python
import pandas as pd
def fun_one():
    df=pd.read_csv("emp.csv",index_col="name")
    df2=df.copy()
    return df2
fun_one()
```

```python
def fun4():
    a=fun_3()
    a=a.sort_values(by='salary',ascending=True)
    df2=a.copy()
    return df2
fun4()
```
